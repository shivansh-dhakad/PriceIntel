```python
def storage_to_gb(x):
    if pd.isna(x):
        return None

    x = str(x).upper().strip()

    if 'TB' in x:
        return float(x.replace('TB', '').strip()) * 1024
    elif 'GB' in x:
        return float(x.replace('GB', '').strip())
    else:
        return float(x)

temp = df['SSD Capacity'].apply(storage_to_gb)

df = df[temp >= 256]
df['SSD Capacity'] = df['SSD Capacity'].replace({
    '256': '256 GB',
    '512': '512 GB'
})
```

```python
def clean_os(os):
    os = str(os).lower()

    if 'windows 10' in os:
        return 'Windows 10'
    elif 'windows 11' in os:
        return 'Windows 11'
    elif 'windows 8' in os:
        return 'Windows 8'
    elif 'macos' in os or 'mac os x' in os:
        return 'macOS'
    elif 'linux' in os or 'ubuntu' in os:
        return 'Linux'
    elif 'dos' in os:
        return 'DOS'
    elif 'chrome' in os:
        return 'Chrome OS'
    else:
        return 'Other'

df['Operating System'] = df['Operating System'].apply(clean_os)
```

```python
def cpu_brand(cpu):
    cpu = str(cpu)

    if 'Intel' in cpu:
        return 'Intel'
    elif 'AMD' in cpu:
        return 'AMD'
    elif 'Apple' in cpu:
        return 'Apple'
    elif 'Qualcomm' in cpu:
        return 'Qualcomm'
    elif 'MediaTek' in cpu:
        return 'MediaTek'
    else:
        return 'Other'

df['CPU_Brand'] = df['Processor'].apply(cpu_brand)
```

```python
df['CPU_Brand'].unique()
```

```python
import re

def cpu_category(cpu):
    cpu = str(cpu)

    # Apple
    if 'M1' in cpu:
        return 'Apple M1'
    elif 'M2' in cpu:
        return 'Apple M2'
    elif 'M3' in cpu:
        return 'Apple M3'
    elif 'M4' in cpu:
        return 'Apple M4'

    # Core Ultra
    ultra = re.search(r'Ultra\s+([579])', cpu)
    if ultra:
        return f'Ultra {ultra.group(1)}'

    # Intel
    intel = re.search(r'i([3579])[-\s]?(\d{4,5})', cpu)

    if intel:
        family = f'i{intel.group(1)}'
        model = intel.group(2)

        if len(model) == 4:      # 8250U, 9750H
            gen = model[0]
        else:                    # 12450H, 13420H
            gen = model[:2]

        return f'{family}-{gen}Gen'

    # AMD
    amd = re.search(r'Ryzen\s+([3579])\s+(\d{4})', cpu)

    if amd:
        family = f'Ryzen {amd.group(1)}'
        gen = amd.group(2)[0]

        return f'{family}-{gen}Gen'

    return 'Other'

df['CPU_Category'] = df['Processor'].apply(cpu_category)
```
